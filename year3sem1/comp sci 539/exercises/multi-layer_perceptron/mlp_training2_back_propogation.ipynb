```python
import numpy as np
```

```python
X = np.array([[1, 2], [2, -1], [3, 0]])
W1 = np.array([[0.1, -0.1, 0.2], [0, 0.3, 0.05]])
W2 = np.array([[-0.1], [0.5], [-0.1]])
y = np.array([[0], [1], [0]])

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# d
U1 = np.dot(np.hstack((np.ones((X.shape[0], 1)), X)), W1.T)
Z1 = sigmoid(U1)
U2 = np.dot(np.hstack((np.ones((Z1.shape[0], 1)), Z1)), W2)
Z2 = sigmoid(U2)

# e ???
delta2 = (Z2 - y) * Z2 * (1 - Z2)
delta1 = np.dot(delta2, W2[:, 1:]) * Z1 * (1 - Z1)

# f
grad_W2 = np.dot(np.hstack((np.ones((Z1.shape[0], 1)), Z1)).T, delta2) / X.shape[0]
grad_W1 = np.dot(np.hstack((np.ones((X.shape[0], 1)), X)).T, delta1) / X.shape[0]

# g
learning_rate = 0.1
W2 -= learning_rate * grad_W2
W1 -= learning_rate * grad_W1
```
